Question no.1

```python
page=requests.get(' https://www.cnbc.com/world/?region=world ')
page
```

```python
soup=BeautifulSoup(page.content,'html')
```

```python
Headings=[]

for heading in soup.find_all('div',class_='LatestNews-headlineWrapper'):
    i=heading.find('a')
    Headings.append(i.text.strip())
```

```python
Date=[]

for date in soup.find_all('time',class_='LatestNews-timestamp'):
    Date.append(date.text.strip())
```

```python
News_link=[]

for link in soup.find_all('a',class_='LatestNews-headline'):
    i=link.get('href')
    News_link.append(i)
```

```python
len(Headings),len(Date),len(News_link)
```

```python
News=pd.DataFrame({'Heading':Headings,'Date':Date,'News_link':News_link})
```

1) Write a python program to display IMDB’s Top rated 100 Indian movies data

'https://www.imdb.com/list/ls056092300/' (i.e. name, rating, year ofrelease) and make data frame.

```python
page=requests.get('https://www.imdb.com/list/ls056092300/')
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
Name=[]
 
for i in soup.find_all('h3',class_='lister-item-header'):
    name=i.find('a')
    Name.append(name.text)
```

```python
Rating=[]
for i in soup.find_all('div',class_='ipl-rating-star small'):
    rating=i.text.replace('\n','')
    Rating.append(rating)
```

```python
Year_of_release=[]

for i in soup.find_all('span',class_='lister-item-year text-muted unbold'):
    Year_of_release.append(i.text.strip())
    
```

```python
len(Name),len(Rating),len(Year_of_release)
```

```python
Film=pd.DataFrame({'Name':Name,'Rating':Rating,'Year_of_release':Year_of_release})
Film
```

```python
soup=BeautifulSoup(page.content,'html')
```

```python
Product_name=[]
i=0

for name in soup.find_all('h2',class_='clr-shade4 h3-p-name   undefined false  '):
    if i>10:
        break
    Product_name.append(name.text.strip())
    i+=1
```

```python
Price=[]
i=0

for price in soup.find_all('div',class_='discountedPriceText clr-p-black   false  '):
    if i>10:
        break
    Price.append(price.text)
    i+=1
```

```python
Image=[]
i=0

for image in soup.find_all('img',class_='productImgTag'):
    if i>10:
        break
    Image.append(image['src'])
    i+=1
```
