Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and

```python
driver=webdriver.Chrome()
driver.get('https://www.naukri.com/')
```

```python
designation=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[1]/div/div/div/div[1]/div/input')
designation.send_keys('Data Analyst')
```

```python
location_filter=driver.find_element(By.XPATH,'/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[5]/div[2]/div[2]/label/p/span[1]')
location_filter.click()
```

```python
salary=driver.find_element(By.XPATH,'/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[5]/div[2]/div[2]/label/p/span[1]')
salary.click()
```

```python
# scraping job title
Jobtitle=[]
i=0

title_tag=driver.find_elements(By.XPATH,'//a[@class="title "]')
for tag in title_tag:
    if i>=10:
        break
    Jobtitle.append(tag.text.strip())
    i+=1

```

```python
#scraping job location
joblocations=[]
i=0

location_tag=driver.find_elements(By.XPATH,'//span[@class="locWdth"]')
for tag in location_tag:
    if i>=10:
        break
    joblocations.append(tag.text.strip())
    i+=1
    
#scraping company_name

companyname=[]
i=0

name_tag=driver.find_elements(By.XPATH,'//span[@class=" comp-dtls-wrap"]//a[1]')
for name in name_tag:
    if i>=10:
        break
    companyname.append(name.text.strip())
    i+=1
    
#scraping experience

experiencerequired=[]
i=0

experience_tag=driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
for tag in experience_tag:
    if i>=10:
        break
    experiencerequired.append(tag.text.strip())
    i+=1
```

```python
len(Jobtitle),len(joblocations),len(companyname),len(experiencerequired)
```

```python
Jobs=pd.DataFrame({'Jobtitle':Jobtitle,'Job_location':joblocations,'Company_name':companyname,'Experience_required':experiencerequired})
Jobs
```

Q8: Write a python program to display list of 50 Most expensive cars in the world
(i.e. Car name and Price) from https://www.motor1.com/
This task will be done in following steps:
1. First get the webpage https://www.motor1.com/
2. Then You have to type in the search bar ’50 most expensive cars’
3. Then click on 50 most expensive cars in the world..
4. Then scrap the mentioned data and make the dataframe.

```python
driver=webdriver.Chrome()
driver.get('https://www.motor1.com/')
time.sleep(1)
```

```python
search=driver.find_element(By.XPATH,'/html/body/div[9]/div[2]/div/div/div[3]/div/div/div/form/input')
search.send_keys('50 most expensive cars')
```

```python
search_button=driver.find_element(By.XPATH,'/html/body/div[9]/div[2]/div/div/div[3]/div/div/div/form/button[1]')
search_button.click()
```

```python
select=driver.find_element(By.XPATH,'/html/body/div[9]/div[9]/div/div[1]/div/div/div[1]/div/div[1]/h3/a')
select.click()
```

```python
car_names=[]

name_tag=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')[0:50]
for tag in name_tag:
    car_names.append(tag.text)
```

```python
prc=[]

price_tag=price_tags = driver.find_elements(By.XPATH,"//strong")

for tag in price_tags:
    result=re.findall('\d+.\d \w+',tag.text.strip())
    prc.append(result)
    
```

```python
len(car_names),len(prc)
```

```python
Car=pd.DataFrame({'car_name':car_names,'Price':prc})
Car
```
